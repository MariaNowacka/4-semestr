```python
import numpy as np, matplotlib.pyplot as plt
```

## Mieszany proces Poissona

```python
def algorytm1(L,T):
    lam = L
    t = 0
    I = 0
    S = []
    while t <= T:
        U = np.random.rand()
        t += -1/lam*np.log(U)
        I += 1
        S.append(t)
    return S, I
```

```python
a1 = [algorytm1(np.random.uniform(0,10), 10)[0] for _ in range(15)]
```

```python
for a in range(15):
    plt.step(a1[a], range(len(a1[a])), where='post')
```

```python
a2 = [algorytm1(np.random.exponential(1), 10)[0] for _ in range(15)]
for a in range(15):
    plt.step(a2[a], range(len(a2[a])), where='post')
    # plt.xlim(0,10)
```

```python
def algorytm2(L,T):
    lam = L
    n = np.random.poisson(lam*T)
    if n == 0:
        break
    U = [np.random.uniform(0, T) for _ in range(n)]
    t = 0
    I = 0
    S = []
    while t <= T:
        U = np.random.rand()
        t += -1/lam*np.log(U)
        I += 1
        S.append(t)
    return S
```

## Mieszany proces Poissona

```python
T = 10
N = [algorytm1(np.random.uniform(0,10), T)[1] for _ in range(200)]
Nhalf = [algorytm1(np.random.uniform(0,10), T/2)[1] for _ in range(200)]
   
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,6))
axes[0].hist(N, bins = 20)
axes[1].hist(Nhalf, bins = 20)
```

```python
Ne = [algorytm1(np.random.exponential(1), T)[1] for _ in range(200)]
Nehalf = [algorytm1(np.random.exponential(1), T/2)[1] for _ in range(200)]
   
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,6))
axes[0].hist(Ne, bins = 20)
axes[1].hist(Nehalf, bins = 20)
```

```python
Nn = [algorytm1(np.random.pareto(3), T)[1] for _ in range(200)]
Nnhalf = [algorytm1(np.random.pareto(3), T/2)[1] for _ in range(200)]
```
