**Generator ACORN (Additive Congruential Random Number)** \
Generator ACORN $k$-tego rzędu oparty jest na ciągach liczbowych zdefiniowanych poprzez równania rekurencyjne
  $$X^0_n  = X^0_{n-1},\ n \geq 1 \\
  X^m_n  = (X^{m-1}_{n} + X^m_{n-1})\operatorname{mod}M, \\ m = 1, \dots,k,\ n\geq 1 \\
  Y^k_n = \frac{X^k_n}{M},\ n \geq 1.$$
Jeśli ziarno (pierwszy element wektora o elementach $X^m_0 ∈ {0 . . . M}$) , czyli $X_0^0 ∈ {1 . . . M − 1}$, jest względnie
pierwsze z $M$ oraz $M$ jest dostatecznie dużą liczbą naturalną, to generator posiada pożądane cechy a ciąg $Y^k_n$ dla
$n ≥ 1$ przypomina ciąg IID zmiennych losowych z rozkładu jednostajnego $U(0, 1)$. Zaimplementuj generator jako
funkcję `ACORN(N, k, M, Lag)` zwracającą `N` liczb pseudolosowych $Y^k_n$, dla $n = 1, . . . , N,$ korzystając z algorytmu
$k$-tego rzędu, przy ominięciu pierwszych `Lag` wyrazów. Przykładowymi parametrami mogą być `M = 289 − 1` (liczba
pierwsza Mersenne’a), `k = 9`, $X^m_
0 = 0$ dla $m \neq 0$ (mogą być dowolne), `Lag` = $10^3$ (czy to na pewno OK?).
Sprawdź poprawność wyniku robiąc wykresy ciągu $Y^k_n$, funkcji $Y^k_{n−1} \mapsto Y^k_n$ oraz histogramy. Porównaj wydajność z
wbudowanym generatorem numpy. Szczegóły oraz odniesienia do literatury znaleźć możemy na stronie autorów algorytmu https://acorn.wikramaratna.org/index.html.

```python
def ACORN(N, k=9, M = 2**89 - 1, Lag=10**3):
  if N<Lag:
    params = input('podaj dobre parametry')
    ACORN (params)
  c = 11001
  X = [[0]*(N+1) for _ in range(k+1)]
  X[0][0] = c
  n = len(X) # wiersze
  m = len(X[0]) # kolumny
  for i in range(n):
    for j in range(1,m):
      X[i][j] = (X[i-1][j] + X[i][j-1])%M
  Yk = [X[k][j]/M for j in range(1,N+1)]
  return Yk[Lag::]
```

```python
a=ACORN(5000)
```

```python
n = np.random.rand(4000)
plt.figure(figsize=(14, 6))
plt.subplot(1,2,1)
plt.hist(a)
plt.title('ACORN')
plt.subplot(1,2,2)
plt.hist(n, alpha = 0.7)
plt.title('numpy')
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1,2,1)
plt.plot(a[0:-1], a[1::])
plt.title('ACORN')
plt.subplot(1,2,2)
plt.plot(n[0:-1], n[1::], alpha = 0.7)
plt.title('numpy')
```
